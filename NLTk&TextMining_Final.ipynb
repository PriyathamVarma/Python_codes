```python
# All Imports

from nltk.corpus import brown
from nltk.book import *
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
required_text_before_capitalization = input()
```

```python
required_text = required_text_before_capitalization
```

# Toeknization - sentence and then word tokenization

```python
required_text_before_sent_token = sent_tokenize(required_text)
```

```python
required_text_before_sent_token
```

```python
sentenes_list = []
for words in required_text_before_sent_token:
  sentenes_list.append(words)
  print(words)
```

```python
sentenes_list
```

```python
selected_sentence = ' '.join(sentenes_list)
selected_sentence
```

#  Word tokenization

```python
words_list = []

wordTokeized = nltk.word_tokenize(selected_sentence)

for words in wordTokeized:
  words_list.append(words)
```

```python
words_list
```

# Stop words removal

```python
# List of stop words
len(stopwords.words('english'))
```

```python
words_after_removal_stopWords = []

for words in words_list:
  if words not in stopwords.words('english'):
    words_after_removal_stopWords.append(words)

    
```

```python
sentence_after_stopWords = ' '.join(words_after_removal_stopWords)
```

# Stemming
# Algorithms used snowballstemming

```python
snowball = SnowballStemmer('english')
```

```python
words_after_stemming = []

for words in words_after_removal_stopWords:
  words_after_stemming.append(snowball.stem(words))
```

```python
words_after_stemming
```

# Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
words_after_lemmatizing = []

for words in words_after_stemming:
  words_after_lemmatizing.append(lemmatizer.lemmatize(words))
```

```python
words_after_lemmatizing
```

# POS(Parts of Speech) Tagging

```python
words_after_posTagged = []

for words in words_after_removal_stopWords:
  words_after_posTagged.append(nltk.pos_tag(nltk.word_tokenize(words)))
```

```python
required_words_after_posTagged = []
for i in range(len(words_after_posTagged)):
  required_words_after_posTagged.append(words_after_posTagged[i][0])
```

```python
required_words_after_posTagged
```

# Named Entity Recognition

```python
# Chunking

words_after_chunked = []

for words in nltk.ne_chunk(required_words_after_posTagged):
  if hasattr(words, 'label'):
    words_after_chunked.append(words.label()+'  '+ ' '.join(c[0] for c in words))
    print(words.label(), ' '.join(c[0] for c in words))
```
