```python
new_dict = {}
for i,value in zip(range(0,new_df.shape[0]),new_df.index):
  print(i)
  new_dict[i] = value
```

```python
new_dict
```

```python
# final dictionary
final_dict = {}
for i,k in zip(range(0,new_df.shape[0]),new_df.index):
  
  final_list = []
  for j,z in zip(range(0,new_df.shape[1]),new_df.index):

    if new_df.iloc[i:i+1,j:j+1].values[0][0] != '1':
      final_list.append(z)
      

  final_dict[k] = final_list
```

```python
final_dict
```

```python
for values in final_dict.values():
  print(values)
```

# combinations

```python
# A Python program to print all
# combinations of given length
from itertools import combinations
mega_list = [] 
# Get all combinations of length 4
# and length 2
for values in final_dict.values():
  #print(values)
  comb = combinations(values, 4)
  print("*************")
  for i in list(comb):
    print(i)
    mega_list.append(i)
```

# Final dataframe

```python
len(mega_list)
```

```python
set_list = set(mega_list)
len(set_list)
```

```python
final_list_after_all = list(set_list)
```

```python
final_list_after_all
```

```python
# converting list of tuples to a dataframe
df = pd.DataFrame.from_records(final_list_after_all, columns=['Policy_1', 'Policy_2', 'Policy_3','Policy_4'])
```
