```python
# Load the data
# Mounting google drive for accessing files
from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
gc = gspread.authorize(GoogleCredentials.get_application_default())
```

```python
# Imports

import pandas as pd
import pandas_datareader as web
```

```python
# Creating a wb variable to store the spreadsheet bu URL

wb = gc.open_by_url('https://docs.google.com/spreadsheets/d/1xFg3zFdHSwK0PDR-VtmHK-7jeOmHg9MEBlNJxliZQC8/edit#gid=1944767530')

# Getting the sheet(if multiple sheets exists) by sheet name

sheet = wb.worksheet('Commercial')

# Getting all values that exist in the sheet

data = sheet.get_all_values()

# Converting the values into a dataframe

data_frame = pd.DataFrame(data)

data_frame
```

```python
# Getting the first row to be the header
data_frame.columns = data_frame.iloc[0]

data_frame = data_frame.iloc[1:].reset_index(drop=True)

new_df = data_frame.set_index('code')
new_df
```

```python
# Get the index name

index_values = new_df.index
index_values
```

```python
# A Python program to print all
# combinations of given length
from itertools import combinations
mega_list = [] 
# Get all combinations of length 4

comb = combinations(index_values, 4)

for i in list(comb):   
    mega_list.append(i)
```

```python
# Creating a list to get all restricted pairs
list_containing_restricted_pairs = []
for i,x in zip(range(0,new_df.shape[0]),new_df.index):

  # List conatining all restrictions pairs taht will be extended to the main list
  helping_list_containing_restrcited_pairs = []
  for j,y in zip(range(0,new_df.shape[1]),new_df.columns):

    if new_df.iloc[i:i+1,j:j+1].values[0][0] == "1":

      z = x,y # Getting both values into one as one argument can be passed to the list 
      helping_list_containing_restrcited_pairs.append(z)

  list_containing_restricted_pairs.extend(helping_list_containing_restrcited_pairs)    
```

```python
len(list_containing_restricted_pairs)
```

```python
final_list_containing_restrictions = []
for values in list_containing_restricted_pairs:
  final_list_containing_restrictions.append(list(values))
```

```python
final_list_without_duplicates = []
for values in final_list_containing_restrictions:
  if values[0] != values[1]:
    final_list_without_duplicates.append(values)
```

# Clearing the restrictions from the main list

```python
final_mega_list = []
for values in mega_list:
  final_mega_list.append(list(values))
```

```python
len(final_mega_list)
```

# Final step

```python
final_mega_list_set = []
for k in final_mega_list:
  final_mega_list_set.append(set(k))
```

```python
restricted_list_set = []
for k in final_list_without_duplicates:
  restricted_list_set.append(list(k))
```

# New way of sorting the combinations

```python
len(final_mega_list_set)
```

```python
# Now we have two lists
# the list containing all entries ----> final_mega_list_set
# the list conatining restricted pairs which cannot be in the final list ---> restricted_list_set
new_list = []
for values in final_mega_list_set:

  new_list2 = []
  test_number = [];

  for values2 in restricted_list_set:

    if values2[0] in values:
      if values2[1] in values:
        test_number.append(-1)
        
      else:
        
        test_number.append(0)
    
  if(sum(test_number) == 0):
    print("the test case")
    print(test_number,values)
    new_list.append(values)
```

# End of the new way

```python
final_number = len(final_mega_list) - len(new_list)
final_number
```

```python
# converting set to list
final_data_list = [list(values) for values in new_list]
```

```python
sorted_list = []
for l in range(0,len(final_data_list)):
  l1 = final_data_list[l]
  l1.sort()

  sorted_list.append(l1)
```

```python
len(sorted_list)
```

```python
id_for_sorted_list = []
for i in range(0,len(sorted_list)):
  id_for_sorted_list.append(i)
```

# Creating the database

```python
# converting list of tuples to a dataframe
df = pd.DataFrame.from_records(sorted_list, columns=['Policy_1', 'Policy_2', 'Policy_3','Policy_4'])
```

```python
df.insert(0,"id",id_for_sorted_list,True)
```

```python
df.to_csv('FINAL-COMMERCIAL-COMBINATIONS.csv')
```
