```python
new_df = data_frame.set_index('code')
new_df
```

```python
# size of dataframe
new_df.shape
```

```python
new_dict = {}
for i,value in zip(range(0,new_df.shape[0]),new_df.index):
  print(i)
  new_dict[i] = value
```

```python
new_dict
```

```python
# final dictionary
final_dict = {}
for i,k in zip(range(0,new_df.shape[0]),new_df.index):
  
  final_list = []
  for j,z in zip(range(0,new_df.shape[1]),new_df.index):

    if new_df.iloc[i:i+1,j:j+1].values[0][0] != '1':
      final_list.append(z)
      

  final_dict[k] = final_list
```

```python
final_dict
```

```python
# for restrictions pairs
# final dictionary
restricted_pairs_dict = {}
final_restricted_pairs_list = []
for i,p in zip(range(0,new_df.shape[0]),new_df.index):
  print("-----------")
  restricted_pairs_list = []
  for j,q in zip(range(0,new_df.shape[1]),new_df.columns):

    if new_df.iloc[i:i+1,j:j+1].values == '1':
      print("********")
      print(p,",",q)
      z= p,q
      restricted_pairs_list.append(z)
      print("********")

  restricted_pairs_dict[i] =   restricted_pairs_list 
  final_restricted_pairs_list.extend(restricted_pairs_list)
```

```python
len(final_restricted_pairs_list)
```

```python
# removing the duplicates from the list
final_restricted_pairs_set = set(final_restricted_pairs_list)
final_restricted_pairs_list_after_cleaning = list(final_restricted_pairs_set)
len(final_restricted_pairs_list_after_cleaning)
```

```python
restricted_pairs_in_list = []
for x in final_restricted_pairs_list_after_cleaning:
  #restricted_pairs_in_list.append(list(x))
  #print(x[0])
  if x[0] != x[1]:
    restricted_pairs_in_list.append(list(x))

```

```python
# A Python program to print all
# combinations of given length
from itertools import combinations
mega_list = [] 
# Get all combinations of length 4
# and length 2
for values in final_dict.values():
  #print(values)
  comb = combinations(values, 4)
  print("*************")
  for i in list(comb):
    print(i)
    mega_list.append(i)
```
