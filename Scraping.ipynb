```python
#WEB SCRAPING

import requests

#Grabing a title

res = requests.get("http://www.civilios.com/")

type(res)
```

```python
res.text
```

```python
#Using beautiful soup

import bs4

soup = bs4.BeautifulSoup(res.text, "lxml")#we use lxml library for BS to convert into a HTML document

soup
```

```python
#Selecting one tag from the HTML

soup.select("p")
```

```python
#Selecting only one out of many

soup.select("p")[0]
```

```python
#Only presenting it in a text format

soup.select("p")[0].getText()
```

```python
#Grabbing images


x = soup.select("img")[0]
```
